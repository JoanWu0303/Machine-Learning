## 1. Loading Package

```python
#Basic import
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 

#Machine Learning
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

## 2. Data Loading

```python
data=pd.read_csv(r'/Users/joan/Desktop/Spring2022/Machine Learning 6364/HW2/baseball-9192.csv')
data.drop(columns= ['Unnamed: 0','Name'],inplace =True)
```

```python
data.head()
```

## 3. Data Convert and Modeling

##### **3.1 Convert the dataframe into array (matrix)**

```python
data_matrix=np.array(data)
```

 ##### **3.2 Split the data into X(features) and Y(target)**

```python
#matraix_name[row, column]
X=data_matrix[:,1:]
Y=data_matrix[:,0]
```

##### 3.3 Split data into two sets: (1) Train data set (2) Test data set

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=40)
```

##### 3.4 Run the LinearRegression Model

```python
model=LinearRegression()
X_train_re = []
X_test_re=[]
ans=[]

for i in np.arange(len(X_train[0])):
    X_train_re = np.array(X_train[:,i]).reshape(-1,1)
    X_test_re = np.array(X_test[:,i]).reshape(-1,1)
    
    Y_train_re=np.array(Y_train).reshape(-1,1)
    Y_test_re=np.array(Y_test).reshape(-1,1)
    
    # Train the model using the training sets
    model.fit(X_train_re, Y_train_re)

    #predit the training set
    model_y_predict = model.predict(X_test_re)
    
    #calculate the results
    score=model.score(X_train_re, Y_train_re)
    coefficients=model.coef_
    MSE=mean_squared_error(Y_test_re, model_y_predict)
    MAE=mean_absolute_error(Y_test_re, model_y_predict)
    RSME=np.sqrt(mean_squared_error(Y_test_re, model_y_predict))
    r2=r2_score(Y_test_re, model_y_predict)
    
    result=[score,coefficients,MSE,MAE,RSME,r2]
    ans.append(result)
    
```

```python
#Drop the column named = Salary
data.drop(columns= ['Salary'],inplace =True)
a = data.columns.tolist()
a
```

##### 3.4 Output the results from the model

```python
type_math=['Score','Coefficients','MSE','MAE','RSME','r2']
df = pd.DataFrame(data=ans,
                  index=a,
                  columns=type_math
                 )
df
```

##### 3.5 Sorting the result values by "r2"

```python
df.sort_values(by='r2', axis=0, ascending=False, inplace=True)
df
```

##### From the above table, we can tell that "RBI" has the highest values on r2 which we can say "RBI" has the best performance in single feature linear regression mode.
